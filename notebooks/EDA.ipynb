```python
# Step 3.1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make charts look nice
sns.set(style="whitegrid")

# Step 3.2: Load the dataset
df = pd.read_csv("/Users/akashbhat/credit-card-fraud-detection/data/raw/creditcard.csv")  # adjust path if needed
print("Shape of dataset:", df.shape)
df.head()
```

```python
# Step 3.3: Fraud distribution
fraud_count = df["Class"].value_counts()
print(fraud_count)

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_count.index, y=fraud_count.values)
plt.xticks([0,1], ["Normal", "Fraud"])
plt.title("Class Distribution (Normal vs Fraud)")
plt.show()
```

```python
# Step 3.4: Amount distribution
plt.figure(figsize=(8,4))
sns.histplot(df["Amount"], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
# Step 3.5: Add hour of transaction
df["Hour"] = (df["Time"] / 3600) % 24

plt.figure(figsize=(10,5))
sns.histplot(data=df, x="Hour", hue="Class", bins=24, element="step", stat="density", common_norm=False)
plt.title("Fraud vs Normal Transactions by Hour")
plt.show()
```
